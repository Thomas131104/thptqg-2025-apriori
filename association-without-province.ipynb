## Tiá»n xá»­ lÃ½ dá»¯ liá»‡u

```python
import polars as pl 
import matplotlib.pyplot as plt

df = pl.scan_csv("output/20250715-ketquathi-ct2018a.csv")

df = df.with_columns([
    pl.col("SOBAODANH").cast(pl.Utf8).str.zfill(8).alias("SOBAODANH"),
    
    *[
    pl.col(col).cast(pl.Float64).alias(col)
        for col in df.collect_schema().names()
        if col not in ["STT", "SOBAODANH", "MÃ£ tá»‰nh thÃ nh"]
    ]
])

df.collect().sample(n=5)
```

```python
df.collect_schema().names()
```

```python
df.describe()
```

```python
necessary_df_bool = df.drop(["STT", "SOBAODANH", "ToÃ¡n", "VÄƒn", "MÃ£ tá»‰nh thÃ nh"])
    
necessary_df_bool = necessary_df_bool.select(
    [pl.col(c).is_not_null().alias(c) for c in necessary_df_bool.collect_schema().names()]
).collect().to_pandas()

necessary_df_bool.head()
```

## Luáº­t káº¿t há»£p (Apriori)

```python
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
import matplotlib.pyplot as plt
```

```python
combine = apriori(
    necessary_df_bool, 
    min_support = 0.01, 
    use_colnames = True
)

rules = association_rules(
    combine, 
    metric="confidence", 
    min_threshold=0.25
)
```

```python
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'leverage']].round(2)
```

## Luáº­t káº¿t há»£p (FPGrowth)

```python
combine = fpgrowth(
    necessary_df_bool, 
    min_support = 0.01, 
    use_colnames = True
)

rules = association_rules(
    combine, 
    metric="confidence", 
    min_threshold=0.25
)
```

```python
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift', 'leverage']].round(2))
```
